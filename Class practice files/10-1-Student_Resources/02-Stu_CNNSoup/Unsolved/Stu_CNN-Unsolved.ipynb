```python
# Read HTML from file
filepath = os.path.join("..", "Resources", "template.html")
with open(filepath) as file:
    html = file.read()
```

```python
# Create a Beautiful Soup object
cnn_soup = bs(html, 'html.parser')
print(cnn_soup.prettify())
```
